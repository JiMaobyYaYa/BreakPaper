```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv(r'newPaperData.csv')
df
```

```python
df_deleteall0 = df.loc[:, ~(df == 0).all()]
# df.set_index('Timestamp', inplace=True) 设置索引
df_deleteall0
```

```python
corr_table=df_deleteall0.corr()
corr_table
```

```python
import seaborn as sns
plt.figure(figsize=(200, 160))  # 设置图形大小  
sns.heatmap(corr_table, annot=True, cmap='coolwarm', fmt=\".2f\")  # annot=True显示数值，cmap设置颜色映射，fmt设置数值格式  
plt.title('Correlation Heatmap')  # 设置标题  
plt.show()"
```
